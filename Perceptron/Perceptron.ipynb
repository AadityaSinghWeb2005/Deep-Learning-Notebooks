```python
#Importing Libraries
import numpy as np
import pandas as pd
from sklearn.linear_model import Perceptron
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_classification
```

```python
#Dataset (Sklearn Make Classification Dataset)
X, y = make_classification(
    n_samples=100,      # total rows
    n_features=2,       # ONLY 2 input columns
    n_redundant=0,
    n_informative=2,
    n_clusters_per_class=1,
    n_classes=2,        # binary output
    random_state=42
)

# Convert to DataFrame
df = pd.DataFrame(X, columns=["Feature1", "Feature2"])
df["Output"] = y
```

```python
#Checking the data
print('Shape :',df.shape)
print('Random Five Rows of Dataset :\n',df.sample(5))
```

```python
#Visualizing the data set
sns.scatterplot(
    x = df['Feature1'],
    y=df['Feature2'],
    hue = df['Output']
)
```

```python
#Dividing the Dataset
x = df.iloc[:,0:-1]
y = df.iloc[:,-1]
```

```python
#Training the Perceptron
P = Perceptron()
P.fit(x,y)
```

```python
#Intercept and Coefficients
print(f'Intercept : {P.intercept_}')
print(f'Coefficients : {P.coef_}')
```
